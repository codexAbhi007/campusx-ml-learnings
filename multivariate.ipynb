```python
import seaborn as sns 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'],style=tips['smoker'],size=tips['size'])
```

```python
titanic = pd.read_csv('train.csv')
```

```python
titanic.head()
```

```python
sns.barplot(x=titanic['Pclass'],y=titanic['Age'])
```

```python
sns.barplot(x=titanic['Pclass'],y=titanic['Fare'],hue=titanic['Sex'])
```

```python
sns.boxplot(x=titanic['Sex'],y=titanic['Age'],hue=titanic['Survived'])
```

```python
sns.kdeplot(
    titanic[titanic['Survived'] == 0]['Age'],
    label='Did Not Survive',
    fill=False
)

sns.kdeplot(
    titanic[titanic['Survived'] == 1]['Age'],
    label='Survived',
    fill=False
)
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
(titanic.groupby('Pclass')['Survived'].mean() * 100).plot(kind='bar')
```

```python
ax = (titanic.groupby('Pclass')['Survived'].mean() * 100).plot(kind='bar')

ax.bar_label(ax.containers[0], fmt='%.1f%%')

plt.ylabel('Survival Rate (%)')
plt.xlabel('Passenger Class')
plt.title('Survival Percentage by Passenger Class')

plt.show()
```

```python
ax = (titanic.groupby('Sex')['Survived'].mean() * 100).plot(kind='bar')

ax.bar_label(ax.containers[0], fmt='%.1f%%')
```

```python
ax = (titanic.groupby('Embarked')['Survived'].mean() * 100).plot(kind='bar')

ax.bar_label(ax.containers[0], fmt='%.1f%%')
```
